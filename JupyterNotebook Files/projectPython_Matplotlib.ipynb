```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Basic line graph using plot() using list 
plt.plot([1, 3, 2], [4, 2, 4])
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60]
y = [13, 45, 23, 34, 96, 76]
plt.title('Bar Graph')
plt.bar(x, y, color='dodgerblue', width=5)
plt.show()
```

```python
# Default x-points values as 0,1,2,3...  
y = np.array([2,4,1,5])

plt.figure()
plt.plot(y)
plt.show()
```

```python
plt.plot(y,'r')
plt.show()
```

```python
plt.plot(y,'o')
plt.show()
```

```python
x = [10,4,7,1]
y = [2,8,3,9]

plt.plot(x, marker = 's')
plt.plot(y, marker = 'D')

plt.show()
```
